```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("pavansubhasht/ibm-hr-analytics-attrition-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(path + '/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='JobRole')
plt.xticks(rotation=15)
plt.title("Employee count by JobRole")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Gender', y='MonthlyIncome',data=df)
plt.title("Avg income by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['MonthlyIncome'].hist(bins=30)
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='JobLevel', y='MonthlyIncome', data=df)
plt.xticks(rotation=20)
plt.title("MonthlyIncome by JobRole")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Gender', y='YearsAtCompany',data=df, hue='MaritalStatus')
plt.title("Avg years at company by Gender and Marital status")
plt.show()
```

```python
sns.countplot(data=df, x='Attrition', hue='OverTime')
```

```python
num_df = df.select_dtypes(include='number')
correlation = num_df.corr()
plt.figure(figsize=(14,10))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Heatmap of Numeric Columns")
plt.show()
```

Observations:

A significantly high number of employees (>300) are in the Sales Executive role, whereas the Human Resources department has the lowest count (<50 employees).

Average monthly income is similar across genders, indicating minimal gender-based income disparity in the dataset.

The majority of employees earn in the lower-to-middle income range. A clear positive relationship exists between Job Level and Monthly Income, reflecting the expected progression in pay with promotions.

Married employees tend to have longer tenures at the company across both genders, while single male employees generally have shorter tenures.

Employees working overtime are much more likely to experience attrition, suggesting that excessive workload may be a key factor driving resignations.

Monthly Income shows a strong positive correlation with both Job Level and Total Working Years, reinforcing the link between experience, seniority, and compensation.

Years at Company is positively correlated with both Years in Current Role and Years with Current Manager, which aligns with expected career development paths.

There is a weak correlation between Age and Attrition, indicating that other factors (such as workload, compensation, or work environment) might have a greater influence on employee turnover.
